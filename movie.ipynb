```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l    
```

```python
movie['genres']=movie['genres'].apply(convert)
```

```python
movie['keywords'][0]
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
movie.head()
```

```python
movie['cast'][0]
```

```python
#selecting first 3 actors
```

```python
def convert2(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
          l.append(i['name'])
          counter+=1
        else:
            break
    return l    
```

```python
movie['cast']=movie['cast'].apply(convert)
```

```python
movie.head()
```

```python
movie['crew'][0]
```

```python
#finding directors name
```

```python
def convert3(obj):
    l=[]
    for i in ast.literal_eval(obj):
          if i['job']=='Director':
            l.append(i['name'])
    return l 
```

```python
movie['crew']=movie['crew'].apply(convert3)
```

```python
movie.head()
```

```python
movie['overview'][0]
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split())
```

```python
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movie.head()
```

```python
movie['tags']=movie['overview']+movie['genres']+movie['keywords']+movie['cast']+movie['crew']
```

```python
movie
```

```python
new_df=movie[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
       y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
#convert text into vectors(bag of words,tfidf)
#in tags which five vectors are nearby we choose for recommend
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer( max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#finding distance from each vectors
sim=cosine_similarity(vectors)
```

```python
sim.shape
```

```python
#similarity of 1st movie with all other 
sim[0]
```

```python
sim[1]
```

```python
sorted(list(enumerate(sim[1])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=sim[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Avatar')
```
